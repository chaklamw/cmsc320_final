```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Features of the dataset
df = pd.read_csv('dataset.csv')
len(df)



#Sorts caffeine intake into categorical data 
df['caffeine_groups'] = pd.cut(df['caffeine_intake_cups'],
                               bins = 3,
                               labels = ['0-1 Cups', '2 Cups', '3-4 Cups'])
list(df.columns)
```

## Some preliminary hypothesis testing


- H0: Caffeine Intake has no impact on sleep quality
- HA: Caffeine Intake has a negative impact on sleep quality

- alpha = 0.05

```python
from scipy import stats

r, p_value = stats.pearsonr(df['caffeine_intake_cups'], df['sleep_quality_score'])
print(f'Correlation: {r:.4f}')
print(f'P-value: {p_value:.4f}')
```

p-value of 0.27 is > alpha so we fail to reject H0 and caffeine intake has no significant impact on sleep quality

```python
plt.scatter(df['caffeine_intake_cups'], df['sleep_quality_score'])
plt.xlabel("Caffeine Intake (Cups)")
plt.ylabel("Sleep quality")
plt.show()
```

With this scatter plot visualization, there is no correlation between caffeine intake and quality of sleep. In all levels of caffeine intake, it seems that generally sleep quality is the same. There are some people that have very poor sleep quality and there are some that have great sleep quality.

- H0: Occupation has no impact on sleep quality  
- HA: Occupation does have an impact on sleep quality

```python
df['occupation'].unique()
df['occupation'].value_counts()
```

```python
manager = df[df['occupation'] == 'Manager']
doctor = df[df['occupation'] == 'Doctor']
swe = df[df['occupation'] == 'Software Engineer']
freelancer = df[df['occupation'] == 'Freelancer']
student = df[df['occupation'] == 'Student']
researcher = df[df['occupation'] == 'Researcher']
designer = df[df['occupation'] == 'Designer']
teacher = df[df['occupation'] == 'Teacher']

from scipy.stats import f_oneway

f_stat, p_value = f_oneway(
    manager['sleep_quality_score'],
    doctor['sleep_quality_score'],
    swe['sleep_quality_score'],
    freelancer['sleep_quality_score'],
    student['sleep_quality_score'],
    researcher['sleep_quality_score'],
    designer['sleep_quality_score'],
    teacher['sleep_quality_score'], )

print(p_value)
```

By p-value analysis alone, we reject the null hypothesis and say that occupation has an effect on sleep quality.

```python
#Graphing
#Taking a look at sleep quality distributions by occupation

groups = [df[df['occupation'] == occ]['sleep_quality_score'] for occ in df['occupation'].unique()]
labels = df['occupation'].unique()

plt.boxplot(groups, tick_labels = labels)
plt.title('Sleep Quality Scores by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Sleep Quality Score on a Scale from 1-10')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from pingouin import anova

anova(data = df, dv = 'sleep_quality_score', between = 'occupation', detailed = True)
```

From the distribution of sleep quality scores by occupation, it is difficult to observe the statistical significance shown by the p-value due to the extremely large sample size (n = 15000) of the dataset. Additionally a more detailed anova test shows a partial eta^2 value (effect size of significance) of 0.0014 which is an extremeley small significant difference.

- H0: Physical activity has no impact on sleep quality
- HA: Physical activity has an impact on sleep quality

- alpha = 0.05

```python
statistic, pvalue = stats.spearmanr(df['physical_activity_minutes'], df['sleep_quality_score'])
print(f"P Value: {pvalue}")

plt.scatter(df['physical_activity_minutes'], df['sleep_quality_score'])
plt.xlabel("Physical activity (Minutes)")
plt.ylabel("Sleep quality")
plt.show()
```

With this scatter plot visualization, there is no correlation between physical activity and quality of sleep. In all levels of physical activity, it seems that generally sleep quality is the same. There are some people that have very poor sleep quality and there are some that have great sleep quality. In addition to that, the p value is greater than 0.05 so we fail to reject the null hypothesis.
